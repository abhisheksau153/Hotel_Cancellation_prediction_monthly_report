```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'hotel_bookings.csv')
```

# To find cancel booking of each hotel type

```python
sns.countplot(x='hotel',data=df,hue='is_canceled')       # it will give total numbers of canceled booking of each hotel type
```

# Analysing month wise cancelation

```python
# Now analysing month wise cancelation hotel
# So we need monthly reservation status  as well as reservation status date then we have to link each month with its number
# by creation a dictionary 
# after we have to check is canceled hotel == true  then plot in graph
```

```python
df['reservation_status_date'] = df['reservation_status_date'].apply(pd.to_datetime)
df['reservation_status_month'] = pd.DatetimeIndex(df['reservation_status_date']).month
look_up = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May',
            6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df['reservation_status_month'] = df['reservation_status_month'].apply(lambda x: look_up[x])

df2 = df.loc[(df['is_canceled'] ==  1)]

plt.figure(figsize=(8,8))
splot = sns.countplot(df2['reservation_status_month'])
for p in splot.patches:
  splot.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.xticks(rotation = 90)
plt.show()
```

# Lets find the higest booking monthly

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
plt.figure(figsize=(8,8))
splot = sns.countplot(df['arrival_date_month'], order = months)
for p in splot.patches:
  splot.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.xticks(rotation = 90)
plt.show()
```

So Higest booking done in AUGUST
